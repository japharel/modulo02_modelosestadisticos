Un alumno se dirige a la biblioteca para solicitar el préstamo de un libro y decide que no puede esperar más de $10$ minutos en ser atendido. Supongamos que el bibliotecario tarda por lo menos $0.5$ minutos en atender a una persona, entonces es razonable proponer una distribución uniforme en el intervalo $[0.5,10]$ para modelar el comportamiento de la variable $X$ que es el tiempo en ser atendido (en  minutos).

- Da la función de densidad y gráfica.
- ¿Cuál es la probabilidad de que el tiempo en ser atendido sea mayor a $5$ minutos pero menor a $8$ minutos?
- ¿Cuál es la esperanza y varianza?
- Calcula la función de distribución y gráficala, y con ella calcula: $\mathbb{P}(2.51\le X \le 7.99)$.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform
from IPython.display import Markdown, display
```

```python
def printmd(text):
    display(Markdown(text))
```

```python
# Parámetros de la distribución uniforme
a = 0.5  # mínimo
b = 10   # máximo
dist = uniform(loc=a, scale=b - a)

# 1. Función de densidad
x_vals = np.linspace(0, 11, 500)
pdf_vals = dist.pdf(x_vals)

printmd("### 1. Función de densidad")
printmd(rf"$f_X(x) = \frac{{1}}{{{b} - {a}}} = \frac{{1}}{{{b - a}}} = {1 / (b - a):.4f}$ para $x \in [{a}, {b}]$")

plt.figure(figsize=(8, 4))
plt.plot(x_vals, pdf_vals, label=r"$f_X(x)$", color='blue')
plt.title("Función de densidad de una distribución uniforme en [0.5, 10]")
plt.xlabel("x")
plt.ylabel(r"$f_X(x)$")
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.show()

# 2. P(5 < X < 8)
p_5_8 = dist.cdf(8) - dist.cdf(5)

printmd("### 2. Probabilidad de que $5 < X < 8$")
printmd(rf"$\mathbb{{P}}(5 < X < 8) = F(8) - F(5) = {dist.cdf(8):.4f} - {dist.cdf(5):.4f} = {p_5_8:.4f}$")

# 3. Esperanza y varianza
esperanza = dist.mean()
varianza = dist.var()

printmd("### 3. Esperanza y varianza")
printmd(rf"$\mathbb{{E}}[X] = \frac{{{a} + {b}}}{{2}} = {esperanza:.2f}$")
printmd(rf"$\mathrm{{Var}}(X) = \frac{{({b} - {a})^2}}{{12}} = {varianza:.2f}$")

# 4. Función de distribución y cálculo de probabilidad
cdf_vals = dist.cdf(x_vals)
p_251_799 = dist.cdf(7.99) - dist.cdf(2.51)

printmd("### 4. Función de distribución y cálculo de probabilidad")
printmd(rf"$\mathbb{{P}}(2.51 \le X \le 7.99) = F(7.99) - F(2.51) = {dist.cdf(7.99):.4f} - {dist.cdf(2.51):.4f} = {p_251_799:.4f}$")

plt.figure(figsize=(8, 4))
plt.plot(x_vals, cdf_vals, label=r"$F_X(x)$", color='green')
plt.title("Función de distribución acumulada (CDF) de X")
plt.xlabel("x")
plt.ylabel(r"$F_X(x)$")
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.show()
```
